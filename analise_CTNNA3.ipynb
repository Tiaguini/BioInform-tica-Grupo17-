```python
def contar_features_genbank(arquivo_genbank):
    # Abrir o arquivo GenBank e analisar o registro
    record = SeqIO.read("CTNNA3.gb", "genbank")

    # Inicializar contadores
    count_gene = 0
    count_source = 0
    count_cds = 0

    # Contar features do tipo "gene", "source" e "CDS"
    for feature in record.features:
        if feature.type == "gene":
            count_gene += 1
        elif feature.type == "source":
            count_source += 1
        elif feature.type == "CDS":
            count_cds += 1

    # Imprimir resultados
    print("Quantidade de features do tipo 'gene':", count_gene)
    print("Quantidade de features do tipo 'source':", count_source)
    print("Quantidade de features do tipo 'CDS':", count_cds)

# Substitua "seu_arquivo.genbank" pelo caminho do seu arquivo GenBank
contar_features_genbank("seu_arquivo.genbank")
```
