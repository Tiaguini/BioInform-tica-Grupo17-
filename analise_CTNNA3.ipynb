Carregar módulos necessários:

```python
from Bio import SeqIO
```

Leitura do gene CTNNA3 no ficheiro genbank respetivo:

```python
record = SeqIO.read("CTNNA3.gb", "genbank")
```

Informação global do gene:

```python
print ('Seq:', record.seq)
print('ID:',record.id)
print('Name:',record.name)
print('Size:',len(record.seq))
print('Description:',record.description)
```

Anotações:

```python
annotations=record.annotations
```

```python
print(annotations)
print("")

fonte= record.annotations["source"]
print ("Fonte: ",fonte) 
print("")

organismo=record.annotations["organism"]
print("Espécie: ",organismo)
print("")

taxonomia=record.annotations["taxonomy"]
print("Taxonomia: ",taxonomia)
print("")

tipo= record.annotations["molecule_type"]
print ("Tipo de molécula: ",tipo)
print("")

topologia=record.annotations["topology"]
print ("Topologia: ",topologia )
print("")

divisao= record.annotations["data_file_division"]
print ("Data file division: ",divisao)
print("")

data=record.annotations["date"]
print ("Data: ",data)
print("")

identificador=record.annotations["accessions"]
print("ID NCBI da sequência nucleotídica: ",identificador)

versao=record.annotations["sequence_version"]
print ("Versão da sequência:",versao)
print("")

palavras=record.annotations["keywords"]
print("Palavras-chave: ",palavras)
print("")
```

Features

```python
feat=record.features
print("As features são: ",feat)
```

```python
print("O número de features é: ",len(record.features))
print("")
print("O tipo e localização das features são: ")
for f in record.features: 
    print(f.type, f.location)
```

```python
featcds=[]

for i in range(len(record.features)):
    if record.features[i].type=="CDS":
        featcds.append(i)
    
print("Índices das sequências codificantes: ",featcds)
```

```python
for k in featcds:
    print("Identificação da proteína: ",record.features[k].qualifiers["product"])
    print("Sequência da proteína: ",record.features[k].qualifiers["translation"])
    print("")
```

Gene

```python
feat=record.features
featgene = [ ]
```

```python
for i in range(len(feat)):
    if feat[i].type == "gene":
        featgene.append(i)
for k in featgene:
    print ('Location:',feat[k].location)
    
print("Numero de features tipo gene: ",len(featgene))
```

Qualifiers:

```python
for k in featgene:
    print("Gene:", feat[k].qualifiers['gene'])
    print('Informação adicional:',feat[k].qualifiers['db_xref'])
```

CDS

```python
featcds = [ ]
```

```python
for i in range(len(feat)):
    if feat[i].type == "CDS":
        featcds.append(i)
for k in featcds:
    print ('Location:',feat[k].location)
```

Qualifiers:

```python
for k in featcds:
    print ('Gene:',feat[k].qualifiers['gene'])
    print ('Gene ID:',feat[k].qualifiers['db_xref'])
    print ('Start codons:',feat[k].qualifiers['codon_start'])
    print ('Note:',feat[k].qualifiers['note'])
    print ('Product:',feat[k].qualifiers['product'])
    print ('Protein ID:',feat[k].qualifiers['protein_id'])
    print ('Translation:',feat[k].qualifiers['translation'])
```

```python
featsource = [ ]
```

```python
for i in range(len(feat)):
    if feat[i].type == "source":
        featsource.append(i)
for k in featsource:
    print ('Location:',feat[k].location)
```

Qualifiers:
