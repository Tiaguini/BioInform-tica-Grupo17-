```python
from Bio.Blast import NCBIXML
from Bio.Blast import NCBIWWW
from Bio import SeqIO
from Bio import Entrez
```

```python
def realizar_blastp_e_imprimir_resultados(sequencia_proteica):
    # Executar busca BLASTp em Swiss-Prot
    resultado_blast = NCBIWWW.qblast("blastp", "swissprot", sequencia_proteica.seq)

    # Analisar resultados
    record = NCBIXML.read(resultado_blast)

    # Imprimir informações sobre sequências homólogas
    print("Número de sequências homólogas encontradas:", len(record.alignments))
    
    for alignment in record.alignments:
        e_value = alignment.hsps[0].expect
        accession = alignment.accession
        length = alignment.hsps[0].align_length

        # Recuperar informações detalhadas da sequência homóloga
        handle = Entrez.efetch(db="protein", id=accession, rettype="fasta", retmode="text")
        seq_record = SeqIO.read(handle, "fasta")
        handle.close()

        # Imprimir informações
        print(f"\nAccession: {accession}")
        print(f"E-value: {e_value}")
        print(f"Length of alignment: {length}")
        print("Sequence:")
        print(seq_record.seq)

# Ler sequência proteica do arquivo FASTA
sequencia_proteica = SeqIO.read("ProtODZ3.fasta", format="fasta")

# Realizar BLASTp e imprimir resultados
realizar_blastp_e_imprimir_resultados(sequencia_proteica)
```
